# **Burger's Equation : Data Driven Solution using a neural network**

In one-dimensional space, the burgers equation along with Dirichlet Boundary conditions reads as

**\begin{align}
         u_t + uu_x - (0.01/\pi)u_{xx} = 0 ,  x\in [-1,1] , t\in [0,1]
    \end{align}**

**\begin{align}
         u(0,x) = -sin({\pi}x)
    \end{align}**
**\begin{align}
         u(t,-1) = u(t,1) = 0
    \end{align}**

```python
# import the packages and setup the system

import numpy as np
import tensorflow as tf
import pandas as pd # though we won't need this very much
import matplotlib.pyplot as plt
import seaborn as sns

# sets the default float-type
tf.keras.backend.set_floatx('float32')

# initialize and set the constants for convenience
pi = tf.constant(np.pi,dtype='float32')
mu = 0.01/pi   # the viscosity

# Define initial conditions
def fun_u_0(x):
  return -tf.sin(pi*x)

# Define boundary condition
def fun_u_b(t,x):
  return tf.zeros((t.shape[0],1),dtype='float32')

# Define the residual
def fun_r(t,x,u,u_t,u_x,u_xx):
  return u_t + u*u_x - mu*u_xx
```

```python

# Set the limits for t and x, Here we create two 1D tensors, ll and ul (lowerlimit and upper limit)
tmax = 1
tmin = 0
xmax = 1
xmin = -1
ll = tf.constant([tmin,xmin],dtype='float32')
ul = tf.constant([tmax,xmax],dtype='float32')

# Set the random seed for reproducible results
tf.random.set_seed(0)
```

```python
# Set the data size for boundary, collocation and initial conditions

N_0 = 50
N_b = 50
N_r = 10000
```

```python
# Now its time to create datasets satisfying the boundary conditions

# INITIAL DATA
# u(0,x) = -sin(pi*x)
# t = 0 and x is sampled randomly with uniform distribution in range -1 to 1
t_0 = tf.zeros((N_0,1),dtype='float32')
x_0 = tf.expand_dims(tf.linspace(ll[1],ul[1], num=N_0), axis=-1)
X_0 = tf.concat([t_0,x_0],axis=1)

# We can also evaluate the initial conditions for u: u(0,x)
u_0 = fun_u_0(x_0)

# BOUNDARY DATA
# u(t,1) = u(t,-1) = 0
# t will be sampled with uniform distribution and x with bernoulli distribution randomly
t_b = tf.random.uniform((N_b,1),ll[0],ul[0],dtype='float32')
x_b = ll[1] + (ul[1] - ll[1]) * tf.keras.backend.random_bernoulli((N_b,1), 0.5, dtype='float32')
X_b = tf.concat([t_b,x_b],axis=1)

# Evaluate the boudary conditions at t_b and x_b
u_b = fun_u_b(t_b,x_b)

# Collocation Points
# Both t and x arre drawn randomly with uniform distribution
t_r = tf.random.uniform((N_r,1),ll[0],ul[0],dtype='float32')
x_r = tf.random.uniform((N_r,1),ll[1],ul[1],dtype='float32')
X_r = tf.concat([t_r,x_r],axis=1)

# We won't calculate the residual yet because we don't know the derivatives
```

```python
# Lets see how our data looks like

# we illustrate the collocation points (red circles) and 
# the positions where the boundary and initial conditions will be enforced (cross marks, color indicates value).

fig = plt.figure(figsize=(9,6))
plt.scatter(t_0, x_0, c=u_0, marker='X', vmin=-1, vmax=1)
plt.scatter(t_b, x_b, c=u_b, marker='X', vmin=-1, vmax=1)
plt.scatter(t_r, x_r, c='r', marker='.', alpha=0.1)
plt.xlabel('$t$')
plt.ylabel('$x$')

plt.title('Positions of collocation points and boundary data');
plt.show()
```

```python
## Lets setup the network architecture

def init__model(num_hiddenlayers=4,num_neurons_perlayer=50,act_fn='tanh'):
    model = tf.keras.Sequential()
    
    # Add input layer
    model.add(tf.keras.Input(2))
    
    # Add scalin layer
    scaling_layer = tf.keras.layers.Lambda(lambda x:2.0*(x-ll)/(ul-ll)-1.0)
    model.add(scaling_layer)
    
    # Add the hidden layers
    for i in range(num_hiddenlayers):
        model.add(tf.keras.layers.Dense(num_neurons_perlayer,activation=
                                       act_fn,kernel_initializer='glorot_normal'))
    
    # Add output layer
    model.add(tf.keras.layers.Dense(1))
    
    return model
```

**Now, We define the loss function of our model.**
**\begin{align}
         J(weights,biases) = \frac{1}{N_r} \sum_{i=1}^{N_r} (u_t + uu_x - (0.01/\pi)u_{xx})^2 + 
         \frac{1}{N_0} \sum_{i=1}^{N_0} (u(0,x_0^i)-(-sin(\pi{x_0^i}))^2
         + \frac{1}{N_b} \sum_{i=1}^{N_b} u(t_b^i,x_b^i)^2
    \end{align}**



```python
## Now we need to compute the residual for our loss function
# for that we need derivatives we use tf.Gradienttape

def get_r(model,X_r):

  with tf.GradientTape(persistent=True) as tape:

    t = X_r[:,0:1]
    x = X_r[:,1:2]

    tape.watch(t)
    tape.watch(x)

    # Find u (output of the model)
    u = model(tf.stack([t[:,0], x[:,0]], axis=1))

    u_t = tape.gradient(u,t)
    u_x = tape.gradient(u,x)
    u_xx = tape.gradient(u_x,x)
  
  del tape
  return fun_r(t,x,u,u_t,u_x,u_xx)
```

```python
## Now lets create the loss function

def get_loss(model,X_r,X_0,X_b,u_0,u_b):

  # Netloss = Initial_data_loss + boundary_data_loss + residual_loss
  r = get_r(model,X_r)
  loss = tf.reduce_mean(tf.square(r))

  # Initial_data loss
  pred_0 = model(X_0)
  loss+= tf.reduce_mean(tf.square(u_0-pred_0))

  # Boundary loss
  pred_b = model(X_b)
  loss+=tf.reduce_mean(tf.square(u_b-pred_b))

  return loss
```

```python
## To calculate the gradient with respect to parameters
def get_grad(model, X_r, X_0,X_b, u_0,u_b):
    
    with tf.GradientTape(persistent=True) as tape:
        # This tape is for derivatives with
        # respect to trainable variables
        tape.watch(model.trainable_variables)
        loss = get_loss(model, X_r,X_0,X_b,u_0,u_b)

    g = tape.gradient(loss, model.trainable_variables)
    del tape

    return loss, g
```

```python
## Set up the optimizer and train the model

model = init__model()

# We choose a piecewise decay of the learning rate, i.e., the
# step size in the gradient descent type algorithm
# the first 1000 steps use a learning rate of 0.01
# from 1000 - 3000: learning rate = 0.001
# from 3000 onwards: learning rate = 0.0005

lr = tf.keras.optimizers.schedules.PiecewiseConstantDecay([1000,3000],[1e-2,1e-3,5e-4])

# Choose the optimizer
optim = tf.keras.optimizers.Adam(learning_rate=lr)
```

```python
# Train the model

from time import time

# Define one training step as a TensorFlow function to increase speed of training
@tf.function
def train_step():
    # Compute current loss and gradient w.r.t. parameters
    loss, grad_theta = get_grad(model, X_r, X_0,X_b, u_0,u_b)
    
    # Perform gradient descent step
    optim.apply_gradients(zip(grad_theta, model.trainable_variables))
    
    return loss

# Number of training epochs
N = 20000
hist = []

# Start timer
t0 = time()

for i in range(N+1):
    
    loss = train_step()
    
    # Append current loss to hist
    hist.append(loss.numpy())
    
    # Output current loss after 50 iterates
    if i%50 == 0:
        print('It {:05d}: loss = {:10.8e}'.format(i,loss))
        
# Print computation time
print('\nComputation time: {} seconds'.format(time()-t0))
```

```python
## Plot the solution

from mpl_toolkits.mplot3d import Axes3D

# Set up meshgrid
N = 600
tspace = np.linspace(ll[0], ul[0], N + 1)
xspace = np.linspace(ll[1], ul[1], N + 1)
T, X = np.meshgrid(tspace, xspace)
Xgrid = np.vstack([T.flatten(),X.flatten()]).T

# Determine predictions of u(t, x)
upred = model(tf.cast(Xgrid,'float32'))

# Reshape upred
U = upred.numpy().reshape(N+1,N+1)

# Surface plot of solution u(t,x)
fig = plt.figure(figsize=(9,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(T, X, U, cmap='viridis');
ax.view_init(35,35)
ax.set_xlabel('$t$')
ax.set_ylabel('$x$')
ax.set_zlabel('$u_\\theta(t,x)$')
ax.set_title('Solution of Burgers equation');
#plt.savefig('Burgers_Solution.pdf', bbox_inches='tight', dpi=300);
```
